```python
import pandas as pd
import seaborn as sns

# Load a sample dataset
df = sns.load_dataset("tips")  # Sample dataset about restaurant tips

# Display the first few rows (optional)
print(df.head())

# Display the data types of each column
print(df.dtypes)
```

```python
df.drop(["sex", "time"], axis=1, inplace=True)
df.describe()

```

```python
floor_counts = df["size"].value_counts().to_frame()
floor_counts
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x="smoker", y="tip", data=df)
plt.title("Tip Distribution by Smoker Status")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting regplot: size vs total_bill
sns.regplot(x="size", y="total_bill", data=df)
plt.title("Correlation between Table Size and Total Bill")
plt.xlabel("Table Size")
plt.ylabel("Total Bill")
plt.show()

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Prepare data
X = df[["size"]]       # sqft_living stand-in
y = df["total_bill"]   # price stand-in

# Fit linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict and calculate R^2
y_pred = model.predict(X)
r2 = r2_score(y, y_pred)

print(f"R^2 score: {r2}")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Features and target
X = df[["size", "tip"]]
y = df["total_bill"]

# Create pipeline
pipe = Pipeline([
    ('scale', StandardScaler()),
    ('poly', PolynomialFeatures(degree=2, include_bias=False)),
    ('model', LinearRegression())
])

# Fit pipeline
pipe.fit(X, y)

# Predict and calculate R^2
y_pred = pipe.predict(X)
r2 = r2_score(y, y_pred)

print(f"R^2 score: {r2}")
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# Define features and target
X = df[["size", "tip"]]
y = df["total_bill"]

# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Create and fit Ridge regression model
ridge_model = Ridge(alpha=0.1)
ridge_model.fit(X_train, y_train)

# Predict and calculate R^2 on test data
y_pred = ridge_model.predict(X_test)
r2 = r2_score(y_test, y_pred)

print(f"R^2 score on test data: {r2}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score

# Define features and target
X = df[["size", "tip"]]
y = df["total_bill"]

# Split into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Polynomial transform (degree 2)
poly = PolynomialFeatures(degree=2, include_bias=False)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

# Ridge regression with alpha=0.1
ridge = Ridge(alpha=0.1)
ridge.fit(X_train_poly, y_train)

# Predict and calculate R^2 on test set
y_pred = ridge.predict(X_test_poly)
r2 = r2_score(y_test, y_pred)

print(f"R^2 score on test data (with 2nd-order polynomial + Ridge): {r2}")
```
